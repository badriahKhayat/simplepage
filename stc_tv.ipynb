```python
import pandas as pd

df = pd.read_excel("/content/stc TV Data Set_T1.xlsb", engine='pyxlsb')
df.head()
```

```python
df.columns

```

```python
# عدد المشاهدين لكل صنف برنامج
df.groupby("program_class")["user_id_maped"].nunique()

```

```python
df['Quality'] = df['hd'].map({0: 'SD', 1: 'HD'})

df[['hd', 'Quality']].head()
```

```python
# عدد المشاهدات لكل مستخدم حسب الجودة
user_quality_counts = df.groupby(['user_id_maped', 'Quality']).size().unstack(fill_value=0)
user_quality_counts.head()
```

```python
def preferred_quality(row):
    if row['SD'] > row['HD']:
        return 'SD Viewer'
    elif row['HD'] > row['SD']:
        return 'HD Viewer'
    else:
        return 'Mixed'

user_quality_counts['Preferred_Quality'] = user_quality_counts.apply(preferred_quality, axis=1)
user_quality_counts.head()
```

```python
user_quality_counts['Preferred_Quality'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Preferred_Quality', data=user_quality_counts)
plt.title("Number of users by quality preference SD or HD")
plt.ylabel("Number of users")
plt.show()
```
